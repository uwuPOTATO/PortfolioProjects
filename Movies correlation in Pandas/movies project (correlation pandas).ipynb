```python
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [12,8]
```

```python
# Read data

df = pd.read_csv('movies.csv', sep=',')
```

```python
df.head()
```

```python
# Check for null values

df.isnull().sum()
```

```python
# See data types

df.dtypes
```

```python
# Create 'YearSplited' column from 'released' column

def fun(x):
    return str((str(x).partition(', ')[2])).partition(' ')[0]


df['YearSplited'] = df.released.apply(fun)

df.head()
```

```python
# Create 'MonthSplited' column from 'released' column

def fun(x):
    return str(x).partition(' ')[0]

df['MonthSplited'] = df.released.apply(fun)

df.head()
```

```python
# Create 'DaySplited' column from 'released' column

def fun(x):
    return str(str(x).partition(' ')[2]).partition(',')[0]

df['DaySplited'] = df.released.apply(fun)

df.head()
```

```python
# Sort values by 'gross'

df = df.sort_values('gross', ascending=False)
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# See if gross depends on budget

plt.scatter(x=df.budget, y=df.gross)

plt.xlabel('Budget')
plt.ylabel('Gross')
plt.title('Gross vs Budget')
```

```python
# Fit the line to the data. Looks like gross depends on budget

sns.regplot(data=df, x=df.budget, y=df.gross, line_kws={'color':'blue'}, scatter_kws={'color':'red'})
```

```python
# Correlation matrix

corr_matrix = df.corr()

corr_matrix
```

```python
# Heatmap

sns.heatmap(corr_matrix, annot=True)


plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.title('Correlation')
```

```python
# Create categorized columns from columns that have 'object' type.
# So we can use heatmap on all features.

df_categorized = df

for col in df_categorized.columns:
    if df_categorized[col].dtype == 'object':
        df_categorized[col] = df_categorized[col].astype('category')
        df_categorized[col] = df_categorized[col].cat.codes

df_categorized.head()
```

```python
sns.heatmap(df_categorized.corr(), annot=True)


plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.title('Correlation')
```

Votes and budget are the only ones above 0.5 correlation with gross
